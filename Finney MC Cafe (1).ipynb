```python
#Richard Finney MC Cafe

menu = { "coffee": {"espresso" : 1.50,
                    "americano" : 2.00,
                    "caffe latte" : 2.25,
                    "mocha" : 2.25,
                    "cappucino" : 2.50,
                    "macchiato" : 2.50},
        "tea" : {"white tea": 1.50,
                "green tea" : 2.00,
                "oolong tea" : 2.25,
                "black tea" : 2.25,
                "earl grey" : 2.50,
                "matcha" : 2.50},
        "pastries" : {"croissant" : 5.50,
                    "cinnamon roll" : 4.50,
                    "banana bread" : 4.25,
                    "pumpkin bread" : 3.75,
                    "sticky bun" : 4.25,
                    "apple turnover" : 3.75}
       }

print("Welcome to MC Cafe!")

def displayMenu():
    print("-" * 50)
    print("Coffee:\n")
    for k,v in menu["coffee"].items():
        print(f"\t{k[0].upper() +k[1:]} : ${v:.2f}\n")
        
    print("Tea:\n")
    for k,v in menu["tea"].items():
        print(f"\t{k[0].upper() +k[1:]} : ${v:.2f}\n")
    
    print("Pastries:\n")
    for k,v in menu["pastries"].items():
        print(f"\t{k[0].upper() +k[1:]} : ${v:.2f}\n")
      
    print("-" * 50)
    
def order():
    cOrd = []
    menu2 = menu.get("coffee")
    menu2.update(menu.get("tea"))
    menu2.update(menu.get("pastries"))
    #print(menu2)
    
    total = 0
    while True:
        try: 
            item = str(input("What would you like to order? ")).lower()
            menu2[item]
            qty = int(input("How many would you like? "))
            price = menu2.get(item)
            #print(price)
            cOrd.append((item, price, qty))
        except:
            print("ERROR: That's not an item on the menu")
        else:    
            o2 = str(input("Would you like to order something else? (y/n) ")).lower()

            if o2 == "y" or o2 == "yes":
                continue
            elif o2 == "n" or o2 =="no":
                print(("-" * 25) + "Receipt" + ("-" * 25))

                for item in cOrd:
                    print(f"{item[2]}\t{item[0]}\t\t\t{(item[1] * item[2]):.2f}")
                    total += (item[1] * item[2])

                print(f"\t\t\tSubtotal: {total:.2f}")
                print(f"\t\t\tSales tax: 6%")
                print(f"\t\t\tTotal: {(total + (total * .06)):.2f}")
                print("-" * 50)
                break
            else:
                print("What? Let's try this again...\n")
        
            

while True:
    see = str(input("Would you like to see the menu?(y/n) ")).lower()
    if see == "y" or see == "yes":
        displayMenu()
        break
    elif see == "n" or see =="no":
        break
    else:
        print("What? Let's try this again...\n")
 

while True:
    o1 = str(input("Would you like to order? (y/n) ")).lower()
    if o1 == "y" or o1 == "yes":
        order()
        break
    elif o1 == "n" or o1 =="no":
        break
    else:
        print("What? Let's try this again...\n")
        

print("Thanks for visiting MC Cafe!")
```
